```python
import pandas as pd
df1=pd.read_csv("student_dataset.csv")
```

```python
print(df1)
```

```python
highest_cgpa=df1.groupby('college_name')['cgpa'].max()
```

```python
highest_cgpa
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
highest_cgpa.plot(kind='bar',color='skyblue')
plt.title('highest cgpa secured by college',fontsize=14)
plt.xlabel('college name',fontsize=12)
plt.ylabel('highest cgpa',fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
highest_cgpa.plot(kind='line',color='skyblue')
plt.title('highest cgpa secured by college',fontsize=14)
plt.xlabel('college name',fontsize=12)
plt.ylabel('highest cgpa',fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
highest_cgpa.plot(kind='barh',color='skyblue')
plt.title('highest cgpa secured by college',fontsize=14)
plt.xlabel('college name',fontsize=12)
plt.ylabel('highest cgpa',fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
highest_cgpa.plot(kind='hist',color='skyblue')
plt.title('highest cgpa secured by college',fontsize=14)
plt.xlabel('college name',fontsize=12)
plt.ylabel('highest cgpa',fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
highest_cgpa.plot(kind='box',color='skyblue')
plt.title('highest cgpa secured by college',fontsize=14)
plt.xlabel('college name',fontsize=12)
plt.ylabel('highest cgpa',fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
highest_cgpa.plot(kind='pie')
plt.title('highest cgpa secured by college',fontsize=14)
plt.xlabel('college name',fontsize=12)
plt.ylabel('highest cgpa',fontsize=12)

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(highest_cgpa,labels=highest_cgpa.index,autopct='%1.1f%%',startangle=140,colors=['skyblue','lightgreen','lightcoral'])
plt.title('percantage distribution of highest cgpa secured by college',fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df1['stud_id'],df1['cgpa'],color='skyblue',edgecolor='black')
plt.title('scatter plot of student id vs cgpa',fontsize=14)
plt.xlabel('student id',fontsize=12)
plt.ylabel('cgpa',fontsize=12)
plt.show()
```
